**การแก้ระบบสมการเชิงเส้นโดยใช้กฏของคราเมอร์ (Use Cramer’ s Rule to solve the system)**  

**สมการ 2 ตัวแปร**  
image.png

```python
import numpy as np
def cramer(matrixA, matrixB): # takes the matrix and the costants

# substitution of the column with costant and creating new matrix
    matrixA1 = np.array([matrixB, matrixA[:,1]]).T
    """สร้าง matrixA1 โดยให้แถวที่หนึ่งของ matrixA1 เป็นคอลัมน์ที่สองของ matrixA (matrixA[:,1])
    และแถวที่สองของ matrixA1 เป็น matrixB โดยใช้ np.array และทำการ transpose (.T) เพื่อเปลี่ยนแถวเป็นคอลัมน์และคอลัมน์เป็นแถว."""
    matrixA2 = np.array([matrixA[:,0], matrixB]).T
    """สร้าง matrixA2 โดยให้แถวที่หนึ่งของ matrixA2 เป็นคอลัมน์แรกของ matrixA (matrixA[:,0])
    และแถวที่สองของ matrixA2 เป็น matrixB โดยใช้ np.array และทำการ transpose (.T) เพื่อเปลี่ยนแถวเป็นคอลัมน์และคอลัมน์เป็นแถว."""
    print("Matrix A :")
    print(matrixA)
    print("Matrix B :")
    print(matrixB)
    print("Matrix A1 :")
    print(matrixA1)
    print("Matrix A2 :")
    print(matrixA2)

#calculatin determinant of the matrix
    detA  = np.linalg.det(matrixA) # calculating the determinant of the original matrix
    detA1 = np.linalg.det(matrixA1)  # calculating the determinant of the new matrix
    detA2 = np.linalg.det(matrixA2)  # calculating the determinant of the new matrix
    print("detA = %.0f" % detA)
    print("detA1 = %.0f" % detA1)
    print("detA2 = %.0f" % detA2)

#finding the X1, X2, X3
    x = detA1/detA
    y = detA2/detA
    print("x = %.2f  และ y = %.2f" %(x,y))

matrixA = np.array([[1,2],
                    [3,4]])

matrixB = np.array([1,2])
cramer(matrixA,matrixB)
```

**การแก้ระบบสมการเชิงเส้นโดยใช้กฏของคราเมอร์ (Use Cramer’ s Rule to solve the system)**  

**สมการ 3 ตัวแปร**   

image.png

```python
import numpy as np
def cramer(matrixA, matrixB): # takes the matrix and the costants

    detA = np.linalg.det(matrixA) # calculating the determinant of the original matrix

# substitution of the column with costant and creating new matrix

    matrixA1 = np.array([matrixB, matrixA[:,1], matrixA[:,2]]).T
    matrixA2 = np.array([matrixA[:,0], matrixB, matrixA[:,2]]).T
    matrixA3 = np.array([matrixA[:,0], matrixA[:,1], matrixB]).T
    print("Matrix A :")
    print(matrixA)
    print("Matrix B :")
    print(matrixB)
    print("Matrix A1 :")
    print(matrixA1)
    print("Matrix A2 :")
    print(matrixA2)
    print("Matrix A3 :")
    print(matrixA3)

#calculatin determinant of the matrix
    detA1 = np.linalg.det(matrixA1)
    detA2 = np.linalg.det(matrixA2)
    detA3 = np.linalg.det(matrixA3)
    print("detA = %.0f" % detA)
    print("detA1 = %.0f" % detA1)
    print("detA2 = %.0f" % detA2)
    print("detA2 = %.0f" % detA3)

#finding the X1, X2, X3
    X1 = detA1/detA
    X2 = detA2/detA
    X3 = detA3/detA

    print("X1 = %.2f  X2 = %.2f  X3 = %.2f" %(X1,X2,X3))

matrixA = np.array([[1, 1, 1],
                    [4, -1, 1],
                    [0, 1, -1]])

matrixB = np.array([6, 5, -1])
cramer(matrixA,matrixB)
```

Calculate profit/loss from revenue and expenses using matrix Subtraction

```python
#ให้นักศึกษาเขียนโปรแกรมคำนวณกำไร/ขาดทุนจากรายได้และค่าใช้จ่ายโดยใช้การลบเมทริกซ์
import numpy as np
revenue = np.array([[180,200,220],
                    [24,36,40],
                    [12,18,20]])
expenses = np.array([[80,90,100],
                     [10,16,20],
                     [8,10,10]])
#เขียนคำสั่งในการหาค่า Profit
profit = revenue - expenses
print("revenue :")
print(revenue)
print("expenses :")
print(expenses)
print("profit :")
print(profit)
```
