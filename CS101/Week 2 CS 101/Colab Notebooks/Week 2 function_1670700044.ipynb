**Scatter Plot และการประยุกต์ใช้ Ordered Pairs**

**Scatter Plot (แผนภูมิจุดกระจาย)**  เป็นกราฟที่ใช้แสดงความสัมพันธ์ระหว่างข้อมูลที่อยู่ในแกน X และ Y โดยการแสดงผลเป็นจุดๆ บนกราฟที่ตำแหน่งของแต่ละจุดถูกกำหนดจากค่าของตัวแปรที่เราอยากจะวิเคราะห์ ความสัมพันธ์ระหว่างตัวแปรทั้งสองนั้นจะแสดงให้เห็นในรูปของจุดที่กระจายอยู่ตามแกน X และ Y

ในการสร้าง scatter plot เราจะใช้ ordered pairs หรือ "คู่ของตัวแปร" ซึ่งหมายถึงการจับคู่ระหว่างค่าของตัวแปรที่อยู่ในแกน X และ Y ในลักษณะ (X, Y) โดยที่ X คือค่าของตัวแปรในแกน X และ Y คือค่าของตัวแปรในแกน Y

**ตัวอย่างเช่น:**  
สมมุติว่าเรามีชุดข้อมูลของ อุณหภูมิ (ที่วัดในระดับเซลเซียส) ในแต่ละวันตลอดทั้งเดือน ค่าของอุณหภูมิในแต่ละวันนั้นสามารถแสดงเป็นคู่ของตัวแปรที่อยู่ในรูปแบบ (วัน, อุณหภูมิ) หรือ (X, Y) เช่น:
- (1, 15) หมายถึงวันที่ 1 ของเดือน และอุณหภูมิ 15 องศาเซลเซียส
- (2, 17) หมายถึงวันที่ 2 ของเดือน และอุณหภูมิ 17 องศาเซลเซียส
- (3, 16) หมายถึงวันที่ 3 ของเดือน และอุณหภูมิ 16 องศาเซลเซียส
- และต่อไปจนถึงข้อมูลครบทุกวันของเดือน

จากชุดข้อมูลคู่อันดับต่อไปนี้จงนำไปใช้ในการ Plot กราฟ   
จากคู่อันดับต่อไปนี้ (1,15),(2,17),(3,16),(4,18),(5,20),(6,20)

````python
#import module
import matplotlib.pyplot as plt

#Create a list for x,y coordinates (Ordered Pairs)
x_list = [1, 2, 3, 4, 5]  # day
y_list = [15, 17, 16, 18, 20]  # Temperature

# Show data in the list of x,y values กำหนด List เก็บค่า
print("x = ",x_list)
print("y = ",y_list)

# Set the X axis to display integer values.
plt.xticks(range(min(x_list), max(x_list)+1))#คำสั่ง หาค่าต่ำสุด สูงสุด

# Add axis details and graph names
plt.title('Temperature vs. Day')
plt.xlabel("Day")# X = วัน
plt.ylabel("Temperature (°C)") # Y = อุณหภูมิ

# Create a scatter plot
plt.scatter(x_list,y_list,s=50, c="blue") #plt.scatter = แสดงค่ารูปแบบของกราฟ , S = ขนาด , C = สี

# Show graph
plt.grid(True)
plt.show()
````

**Using Line Plot**   
**Line Plot (กราฟเส้น)** เป็นเครื่องมือที่ใช้ในการแสดงค่าของตัวแปรที่ต้องการวิเคราะห์ในลักษณะของจุดที่เชื่อมโยงกันด้วยเส้นตรง โดยที่แกน X แสดงค่าของตัวแปรที่เป็นอินพุต (input) และแกน Y แสดงค่าของตัวแปรที่เป็นเอาต์พุต (output) ซึ่งแต่ละจุดบนกราฟจะถูกกำหนดโดยคู่ของค่า (ordered pair) ที่ประกอบไปด้วยค่า x และค่า y ที่สอดคล้องกัน

**การใช้ ordered pair** ในกราฟเส้นนี้จะช่วยให้เราสามารถแสดงข้อมูลที่เชื่อมโยงระหว่างตัวแปร x และ y ที่มีความสัมพันธ์กันอย่างชัดเจน เมื่อจุดต่าง ๆ บนกราฟถูกเชื่อมโยงกันด้วยเส้น จะช่วยให้เราเห็นแนวโน้ม (trend) หรือรูปแบบของการเปลี่ยนแปลงของข้อมูลได้ดียิ่งขึ้น

Example: การเจริญเติบโตของพืชตามเวลา
สมมุติว่าเรามีการศึกษาการเจริญเติบโตของพืชในช่วงเวลาต่าง ๆ โดยการวัดความสูงของพืชในหน่วยเซนติเมตรในแต่ละวัน ตามชุดข้อมูลคู่อันดับ โดยมีความสัมพันธ์จำนวน
วัน (x)	และความสูงของพืช (y) ดังนี้ (0,1.0),(1,1.5),(2,2.1),(3,3.2),(
4,4.0),(5,5.0)  
จากชุดคู่อันดับนี้ เราสามารถสร้างกราฟเส้นที่แสดงความสัมพันธ์ระหว่างจำนวนวันและความสูงของพืช ซึ่งจะช่วยให้เราเห็นแนวโน้มการเจริญเติบโตของพืชตามเวลาได้

**ตัวอย่างการเขียนโปรแกรมสร้างกราฟของยอดขายสินค้า 2 ชนิด ตั้งแต่เดือน มกราคม - พฤษภาคม**  
กำหนดข้อมูลยอดขายสินค้าสองประเภท (Product 1 และ Product 2) ในช่วงเวลา 5 เดือนแรกของปี (มกราคม - พฤษภาคม) และแสดงผลยอดขายในรูปแบบ Line Plot พร้อมกับ Marker เพื่อให้เห็นค่าต่าง ๆ ของยอดขายในแต่ละเดือน

ชุดข้อมูล:  
Product 1: [40, 52, 14, 56, 44]  
Product 2: [42, 55, 64, 82, 93]  
เดือน: ["Jan", "Feb", "Mar", "Apr", "May"]  

````python
import matplotlib.pyplot as plt #import กราฟเส้น matplotlib.pyplot อ้างอง Plt

# Create sample data ข้อมูล
product1 = [40, 52, 14, 56, 44]  # Product sales 1
product2 = [42, 55, 64, 82, 93]  # Product sales 2
month = ["Jan", "Feb", "Mar", "Apr", "May"]  # Months of product sales

# Set the size of the graph
plt.figure(figsize=(9, 6))

# Add axis details and graph names
#กำหนดแกนเพื่อแสดงค่า
plt.title('Sales Information of Product 1 and Product 2')  # Set the graph name
plt.xlabel('Month')  # Set a label for the x-axis
plt.ylabel('Sales')  # Set a label for the y-axis

# Create line plots and marker
plt.plot(
    month, product1,             # x-axis (months) and y-axis (sales of Product 1)
    label='Product 1',           # Label for Product 1 (used in the legend)
    color='black',                # Line color for Product 1
    linestyle='--',              # Line style (dashed line)
    linewidth=2,                 # Thickness of the line
    marker='o',                  # Marker style (circle for each data point)
    markerfacecolor='red',       # Fill color for the markers (red)
    markersize=8                 # Size of the markers
)  # Plot the sales curve for Product 1

plt.plot(
    month, product2,             # x-axis (months) and y-axis (sales of Product 2)
    label='Product 2',           # Label for Product 2 (used in the legend)
    color='black',               # Line color for Product 2
    linestyle=':',               # Line style (dotted line)
    linewidth=2,                 # Thickness of the line
    marker='^',                  # Marker style (triangle for each data point)
    markerfacecolor='orange',    # Fill color for the markers (orange)
    markersize=8                 # Size of the markers
)  # Plot the sales curve for Product 2

# Add legend to show label of data
plt.legend()

# Show the graph
plt.show()
````

#การแก้สมการเพื่อหาค่าตัวแปร x , y โดยใช้

**คำสั่งหลักใน SymPy:**
1. solve: ฟังก์ชันนี้ใช้ในการแก้สมการพีชคณิต สามารถหาค่าของตัวแปรที่ทำให้สมการเป็นจริง เพื่อหาค่าของตัวแปร เช่น x และ y
2. symbols('x'): ใช้ในการสร้างตัวแปร x (หรือตัวแปรอื่น ๆ) ในรูปแบบของสัญลักษณ์ที่สามารถนำไปใช้ในการคำนวณทางคณิตศาสตร์ได้
3. Eq(x**2 - 4, 0): ใช้ในการสร้างสมการ เช่น  เช่น  x^2 - 4 = 0 ซึ่งสามารถใช้ในการหาค่าของ x ที่ทำให้สมการนี้เป็นจริง
4. parse_expr: ใช้ในการแปลงข้อความ (string) ที่ผู้ใช้กรอกให้เป็นนิพจน์ทางคณิตศาสตร์ที่ SymPy สามารถคำนวณได้
5. subs: ใช้แทนค่าของตัวแปรในสมการหรือฟังก์ชันด้วยค่าที่กำหนด


จงเขียนโปรแกรมเพื่อหาค่า x และ y จากสมการที่กำหนดให้

```
2x + y = 6
3x - y = 24
```

````python
from sympy import * #import Sympy เพื่อให้ดึงค่าคำสั่งทั้งหมด

# Defining Symbols
x, y = symbols('x y')

# Create equations
equation1 = Eq(2*x + y, 6)
equation2 = Eq(3*x - y, 24)

# Use solve to find the value of x,y that makes the equation true.
solution = solve((equation1, equation2), (x, y))
print(solution)
# Explanation of the result in 'solution':
# It will be in the form of a dictionary where:
#   - The keys are the variables (x and y).
#   - The values are the numerical solutions for those variables.
#   - The solve function computes the values of x and y that satisfy the given equations.
#   - The result is stored as a dictionary with the format {variable: value}

# Show results
print(f"From the equations \n{equation1}\n{equation2}")
print(f"The solution of the system is x = {solution[x]}, y = {solution[y]}") #แสดงค่า Key X , Key Y
````

**การแทนค่าตัวแปรเพื่อหาผลลัพธ์สมการฟังก์ชันที่โจทย์กำหนด**  
คำสั่งนี้ทำการนำเข้าไลบรารี sympy ซึ่งเป็นไลบรารีที่ใช้ในการคำนวณทางคณิตศาสตร์เชิงสัญลักษณ์ใน Python
- symbols ใช้ในการสร้างสัญลักษณ์ที่เป็นตัวแปร (ในกรณีนี้คือ x)
- parse_expr ใช้ในการแปลงข้อความ (string) ที่ผู้ใช้กรอกเป็นนิพจน์ทางคณิตศาสตร์ที่สามารถคำนวณได้
- subs ใช้แทนค่าของตัวแปรในสมการด้วยค่าที่กำหนด
```
จาก f(x) = 2 * x ** 3 + 5 * x จงหาค่า f(1)  
f(1) = 2* (1) ** 3 + 5 * (1) = 7
```

````python
from sympy import *

# Defining Symbols
x = symbols('x')

# Get the function from the user
f_x = parse_expr(input("Enter the function f(x): "))  # Example: 2*x**3 + 5*x

# Get the value of x from the user
val_x = int(input("Enter the value of x: "))

# .subs is used to substitute the variable value in the function
solution = f_x.subs(x, val_x) #เก็บผลลัพธ์ = ตัวแปร.แทนค่าในสมการ(x, val_x)

# Display the result
print(f"From the equation f(x) = {f_x}, when x = {val_x}, the result is {solution}")
print(f"f({val_x}) = {solution}")
````

**การแทนค่าตัวแปรเพื่อหาผลลัพธ์สมการฟังก์ชันที่โจทย์กำหนด**


การใช้ subs จะเป็นวิธีของ SymPy ที่ใช้แทนค่าของตัวแปรในสมการด้วยค่าที่กำหนด
```
จาก f(x) = 2 * x + 5  เมื่อ x < 2
    f(x) = 4          เมื่อ x >= 2  

จงหาค่า f(-4) ,f(3) ,f(5)  

```  
**ตัวอย่างการแสดงผลลัพธ์**

```
When x =  -4, f( -4) =  2*x + 5, the result is -3
When x =   3, f(  3) =  4, the result is 4
When x =   5, f(  5) =  4, the result is 4
```

````python
from sympy import *

# Defining Symbols
x = symbols('x')

# Create a list of x values
x_list = [-4, 3, 5]

for val_x in x_list:
    # Define the equation based on the condition
    if val_x < 2:
        f_x = 2 * x + 5
        # .subs is used to substitute variable values to find the result
        solution = f_x.subs(x, val_x)
    else:
        f_x = 4
        solution = 4

    # Display the results
    print(f"When x = {val_x:3}, f({val_x:3}) =  {f_x}, the result is {solution}")
````

```
f( -2) = x**2 + 4*x - 1 = -5
f( -1) = x**2 + 4*x - 1 = -4
f(  0) = x**2 + 4*x - 1 = -1
f(  1) = x**2 + 4*x - 1 = 4
f(  2) = x**2 + 4*x - 1 = 11
```





````python
from sympy import *

# Defining Symbols
x = symbols('x')

# Get the function from the user
f_x = parse_expr(input("Enter the function f(x): "))  # x ** 2 + 4 * x - 1

# Get list x
x_list = [-2,-1,0,1,2]

for val_x in x_list :
  # .subs is used to substitute the variable value in the function
  solution = f_x.subs(x, val_x) #เก็บผลลัพธ์ = ตัวแปร.แทนค่าในสมการ(x, val_x)

  # Display the result
  print(f"f({val_x:3}) = {f_x} = {solution}")
````

```
    f(x) = x + 1        เมื่อ x >= 4

จงหาค่า f(0) ,f(4) ,f(-2) ,f(-1) ,f(-3) ,f(10)
```  

ตัวอย่างการแสดงผลลัพธ์  

```
f( 0) = 4*x + 2 = 2
f( 4) = x + 1 = 5
f(-2) = 4*x + 2 = -6
f(-1) = 4*x + 2 = -2
f(-3) = x**2 = 9
f(10) = x + 1 = 11
```



````python
from sympy import *

# Defining Symbols
x = symbols('x')

# Get the function from the user
#f_x = parse_expr(input("Enter the function f(x): "))  # x ** 2 + 4 * x - 1

# Get list x
x_list = [0,4,-2,-1,-3,10]

for val_x in x_list :
  # .subs is used to substitute the variable value in the function
  if val_x < -2 :
    f_x = x ** 2
  elif -2 <= val_x < 4:
    f_x = 4 * x + 2
  elif val_x >= 4:
    f_x = x + 1
  solution = f_x.subs(x, val_x) #เก็บผลลัพธ์ = ตัวแปร.แทนค่าในสมการ(x, val_x)

  # Display the result
  print(f"f({val_x:3}) = {f_x} = {solution}")
````

**Class Activity**  
กำหนด f(x) = 2 * x + 5  
กำหนดค่า x = [-4,-2,0,3,5,6,9]  
จงแทนค่า x ในสมการ เพื่อหาค่า y และนำค่า (x,y) ที่ได้มา Plot Scatter and line ให้ได้ผลลัพธ์ดังตัวอย่าง  
กรอกสมการฟังก์ชัน : f(x): 2 * x + 5  
x =  [-4, -2, 0, 3, 5, 6, 9]  
y =  [-3, 1, 5, 11, 15, 17, 23]  

**ตัวอย่างการแสดงผลลัพธ์**

image.png

````python
from sympy import *
import matplotlib.pyplot as plt

x = symbols('x')

f_x = parse_expr(input("กรอกสมการฟังก์ชัน : ")) #10 2 * x + 5
name = f_x
x_list = [-4, -2, 0, 3, 5, 6, 9]
y_list = []

for val_x in x_list :
  solution = f_x.subs(x, val_x)
  y_list.append(solution)

print("x_list = ",x_list)
print("y_list = ",y_list)

plt.title ("Graph of f(x)")
plt.xlabel('x')
plt.ylabel('y')

# Create a line plot
plt.plot(x_list,
         y_list,
         label=name,
         color='red',
         linestyle='-',
         linewidth=2,
         marker="o",
         markerfacecolor="blue",
         markersize = 10)
plt.grid(True)
plt.legend()
plt.show()
````

````python
from sympy import *
import matplotlib.pyplot as plt

x = symbols('x')

f_x = parse_expr(input("กรอกสมการฟังก์ชัน : ")) #  4*x**2 - 2*x
name = f_x
x_list = [-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10]
y_list = []

for val_x in x_list :
  solution = f_x.subs(x, val_x)
  y_list.append(solution)

print("x_list = ",x_list)
print("y_list = ",y_list)

plt.title ("Graph of f(x)")
plt.xlabel('x')
plt.ylabel('y')

# Create a line plot
plt.plot(x_list,
         y_list,
         label=name,
         color='green',
         linestyle='--',
         linewidth=2,
         marker="o",)
plt.grid(True)
plt.legend()
plt.show()
````
