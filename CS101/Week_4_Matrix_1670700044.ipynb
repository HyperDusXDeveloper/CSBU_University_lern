**เมตริกซ์เอกลักษณ์ (Identity Matrix)**   คือ
เมตริกซ์ที่สมาชิกเส้นทแยงมุมมีค่าเป็น 1 และสมาชิกที่เหลือมีค่าเป็น 0   


```
รูปแบบคำสั่ง np.identity(N, dtype=<class 'float'>)   
N: ขนาดของเมทริกซ์ (จำเป็น)   
dtype: ชนิดของข้อมูลของเมทริกซ์ (default เป็น float)
```



````python
import numpy as np
identity_matrix = np.identity(4) #เป็นการสร้างเมทริกซ์เอกลักษณ์ ที่มีขนาดแบบ 4x4 สมาชิกเส้นทแยงมุมเป็น 1 ทุกตัว ตัวอื่นมีค่าเป็น 0 เป็นเมตริกซ์จัตุรัส ที่มีแถวกับ คอลัมน์ เท่ากัน dtype ชนิดข้อมูลdefault เป็น float
print("Identity Matrix (4x4):")
print(identity_matrix)
````

````python
import numpy as np
identity_matrix = np.identity(4,dtype='int') #เป็นการสร้าง Array 1 มิติ แบบเอกลักษณ์ ที่มีขนาดแบบ 4x4 สมาชิกเส้นทแยงมุมเป็น 1 ทุกตัว ตัวอื่นมีค่าเป็น 0 มี data type เป็น int
print("Integer Identity Matrix (4x4):")
print(identity_matrix)
````

**ทรานสโพสของเมทริกซ์ (Transpose Matrix)**

image.png

````python
import numpy as np
matrixA = np.array([[1,0,5],
                    [-3,7,2]])
print("Matrix A :")
print(matrixA)
#transpose สลับแถวเป็นหลัก สลับหลักเป็นแถว
print("Transpose Matrix A :")
print(matrixA.transpose())
````

# Matrix Operations In Python   
**การบวก ลบ คูณ หาร เมทริกซ์**

**การบวก ลบ เมทริกซ์**  
image.png

````python
import numpy as np
matrixA = np.array([[1,2,3],
                    [0,1,4]])
matrixB = np.array([[2,3,0],
                    [-1,2,5]])
print("Matrix A :")
print(matrixA)
print("Matrix B : ")
print(matrixB)
print("Matrix A + Matrix B: ")
print(matrixA + matrixB)
print("Matrix A - Matrix B: ")
print(matrixA - matrixB)
````

**การคูณเมทริกซ์ด้วยสเกลาร์ (Scalar Multiplication)**   

image.png

````python
import numpy as np
matrixA = np.array([[2,-2,1],
                    [1,-4,0]])
print("Matrix A :")
print(matrixA)
print("Matrix -3A : ")
print(-3*matrixA)
````

**การคูณเมทริกซ์ (Matrix Multiplication)**

image.png



````python
import numpy as np
matrixA = np.array([[3,2],
                    [0,1],
                    [4,2]])
matrixB = np.array([[4,2,6,0],
                    [5,-1,2,1]])
print("Matrix A :")
print(matrixA)
print("Matrix B :")
print(matrixA)
 #สามารถนำอาเรย์มาคูณกันในลักษณะที่เหมือนเป็นการคูณเมทริกซ์ได้โดยใช้ฟังก์ชัน np.dot
matrixAB = np.dot(matrixA,matrixB)
print("Matrix AB = ")
print(matrixAB)
````

**ดีเทอร์มิแนนต์ (Determinant ; det)**  
image.png

````python
import numpy as np
matrixA = np.array([[1,2],
                    [4,6]])
print("Matrix A :")
print(matrixA)
detA = np.linalg.det(matrixA)
print("detA = %.0f"%detA)
````
